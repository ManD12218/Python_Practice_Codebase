## 1. Mean (Average)

### What is Mean?
The **mean** is the average value of all data points. It's calculated by adding all values and dividing by the number of values.

**Formula:** Mean = Sum of all values / Number of values

### What we are going to do:
Calculate the average ad spend across all dentsu campaigns in our dataset.

### Why we are doing it:
Understanding the average spend helps dentsu plan budgets and compare campaign performance across different channels and regions.

```python
# METHOD 1: Calculate mean using pandas .mean() function
mean_spend = df['Ad_Spend_USD'].mean()

# METHOD 2: Calculate mean manually (to understand the formula)
total_spend = df['Ad_Spend_USD'].sum()           # Step 1: Add all values
count = df['Ad_Spend_USD'].count()               # Step 2: Count how many values
mean_spend_manual = total_spend / count          # Step 3: Divide sum by count

print("MEAN AD SPEND ANALYSIS")
print("=" * 50)
print(f"\nTotal Campaigns Analyzed: {count}")
print(f"Total Ad Spend: ${total_spend:,.2f}")
print(f"\nMean (using .mean()): ${mean_spend:,.2f}")
print(f"Mean (manual calculation): ${mean_spend_manual:,.2f}")
```

### Code Explanation:
- `df['Ad_Spend_USD'].mean()` - Calculates the average of the Ad_Spend_USD column
- `df['Ad_Spend_USD'].sum()` - Adds up all values in the column
- `df['Ad_Spend_USD'].count()` - Counts how many non-empty values exist


### Output Explanation:
The mean ad spend tells us that, on average, dentsu spends approximately $30,000 per campaign. Video campaigns have the highest average spend, while Display campaigns have the lowest.

## 2. Median

### What is Median?
The **median** is the middle value when data is arranged in order. If there's an even number of values, it's the average of the two middle values.

### What we are going to do:
Find the median conversions from our dentsu campaign data.

### Why we are doing it:
The median is useful when we have outliers (extreme values). Unlike mean, median isn't affected by very high or very low conversion campaigns.

```python
# METHOD 1: Calculate median using pandas .median() function
median_conversions = df['Conversions'].median()

# METHOD 2: Understanding median step by step
sorted_conversions = df['Conversions'].dropna().sort_values()  # Sort values
n = len(sorted_conversions)                                      # Count values
middle_position = n // 2                                         # Find middle position

# Also calculate mean for comparison
mean_conversions = df['Conversions'].mean()

print("MEDIAN CONVERSIONS ANALYSIS")
print("=" * 50)
print(f"\nTotal Campaigns: {n}")
print(f"Min Conversions: {df['Conversions'].min()}")
print(f"Max Conversions: {df['Conversions'].max()}")
print(f"\nMedian Conversions: {median_conversions}")
print(f"Mean Conversions: {mean_conversions:.2f}")
print(f"\nDifference (Mean - Median): {mean_conversions - median_conversions:.2f}")

```

### Code Explanation:
- `df['Conversions'].median()` - Finds the middle value of conversions
- `sort_values()` - Arranges values from smallest to largest
- `dropna()` - Removes missing (empty) values
- If mean > median, it suggests some high values are pulling the average up

### Output Explanation:
The median provides a reliable measure of typical campaign conversions. Half of all campaigns have conversions below this number, and half have conversions above.

## 4. Range

### What is Range?
The **range** is the difference between the maximum and minimum values in a dataset. It shows the spread of data.

**Formula:** Range = Maximum Value - Minimum Value

### What we are going to do:
Calculate the range of Click-Through Rates (CTR) across dentsu campaigns.

### Why we are doing it:
Understanding the range helps identify the performance gap between best and worst performing campaigns.

```python
# First, let's calculate CTR (Click-Through Rate)
# CTR Formula: CTR = (Clicks / Impressions) × 100

# Create a clean dataset (remove rows with missing values)
df_clean = df.dropna(subset=['Clicks', 'Impressions'])
df_clean = df_clean[df_clean['Impressions'] > 0]  # Avoid division by zero

# Calculate CTR for each campaign
df_clean['CTR'] = (df_clean['Clicks'] / df_clean['Impressions']) * 100

# Calculate range using min() and max()
min_ctr = df_clean['CTR'].min()
max_ctr = df_clean['CTR'].max()
range_ctr = max_ctr - min_ctr

print("RANGE ANALYSIS: CLICK-THROUGH RATE (CTR)")
print("=" * 50)
print(f"\nTotal Campaigns Analyzed: {len(df_clean)}")
print(f"\nMinimum CTR: {min_ctr:.2f}%")
print(f"Maximum CTR: {max_ctr:.2f}%")
print(f"\nRange = Max - Min = {max_ctr:.2f}% - {min_ctr:.2f}% = {range_ctr:.2f}%")
```

### Code Explanation:
- `df['column'].min()` - Finds the smallest value in a column
- `df['column'].max()` - Finds the largest value in a column
- `df.dropna(subset=['col1', 'col2'])` - Removes rows with missing values in specific columns
- `df[df['column'] > 0]` - Filters rows where column value is greater than 0

### Output Explanation:
The range shows significant variation in campaign CTR performance. Search campaigns typically show higher CTR, while Display campaigns have lower but more consistent CTRs.

## 5. Variance

### What is Variance?
**Variance** measures how spread out the data is from the mean. A high variance means data points are far from the mean; a low variance means they are close.

**Formula:** Variance = Sum of (each value - mean)² / (number of values - 1)

### What we are going to do:
Calculate the variance in ad spend across dentsu campaigns.

### Why we are doing it:
Variance helps understand spending consistency. Low variance indicates stable, predictable budgets across campaigns.

```python
# METHOD 1: Calculate variance using pandas .var() function
variance_spend = df['Ad_Spend_USD'].var()

# METHOD 2: Calculate variance manually (to understand the formula)
ad_spend_data = df['Ad_Spend_USD'].dropna()
mean_spend = ad_spend_data.mean()                          # Step 1: Calculate mean
squared_diff = (ad_spend_data - mean_spend) ** 2           # Step 2: (value - mean)²
variance_manual = squared_diff.sum() / (len(ad_spend_data) - 1)  # Step 3: Sum / (n-1)

print("VARIANCE ANALYSIS: AD SPEND")
print("=" * 50)
print(f"\nTotal Campaigns: {len(ad_spend_data)}")
print(f"Mean Ad Spend: ${mean_spend:,.2f}")
print(f"\nVariance (using .var()): {variance_spend:,.2f}")
print(f"Variance (manual calculation): {variance_manual:,.2f}")

```

### Code Explanation:
- `df['column'].var()` - Calculates variance (by default uses n-1 in denominator, which is sample variance)
- `(ad_spend_data - mean_spend) ** 2` - Squares the difference from mean for each value
- We divide by (n-1) instead of n for **sample variance** (more accurate for samples)

### Output Explanation:
The variance values help understand budget consistency. Higher variance means more variability in spending, while lower variance means more consistent budgets.

---

## 6. Standard Deviation

### What is Standard Deviation?
**Standard Deviation** is the square root of variance. It's expressed in the same units as the original data, making it easier to interpret than variance.

**Formula:** Standard Deviation = √Variance

### What we are going to do:
Calculate the standard deviation of conversions across dentsu campaigns.

### Why we are doing it:
Standard deviation helps understand how much campaign conversions vary from the average performance.

```python
# METHOD 1: Calculate standard deviation using pandas .std() function
std_conversions = df['Conversions'].std()

# METHOD 2: Calculate from variance (std = square root of variance)
variance_conversions = df['Conversions'].var()
std_from_variance = np.sqrt(variance_conversions)  # np.sqrt() = square root

# Calculate mean for interpretation
mean_conversions = df['Conversions'].mean()

print("STANDARD DEVIATION ANALYSIS: CONVERSIONS")
print("=" * 50)
print(f"\nTotal Campaigns: {df['Conversions'].count()}")
print(f"Mean Conversions: {mean_conversions:.2f}")
print(f"Variance: {variance_conversions:.2f}")
print(f"\nStandard Deviation (using .std()): {std_conversions:.2f}")
print(f"Standard Deviation (√variance): {std_from_variance:.2f}")

# Interpretation: Most data falls within mean ± 1 std dev
lower_bound = mean_conversions - std_conversions
upper_bound = mean_conversions + std_conversions
print(f"\n📊 INTERPRETATION:")
print(f"Most campaigns (~68%) have conversions between:")
print(f"{max(0, lower_bound):.0f} and {upper_bound:.0f}")

```

### Code Explanation:
- `df['column'].std()` - Calculates standard deviation
- `np.sqrt(value)` - Calculates square root of a number
- Mean ± 1 standard deviation covers approximately 68% of data in a normal distribution

### Output Explanation:
The standard deviation shows how much conversion performance varies across campaigns. A high standard deviation indicates significant variability in performance.

## 7. Quartiles

### What are Quartiles?
**Quartiles** divide data into four equal parts:
- **Q1 (25th percentile)**: 25% of data falls below this value
- **Q2 (50th percentile)**: This is the median - 50% below, 50% above
- **Q3 (75th percentile)**: 75% of data falls below this value
- **IQR (Interquartile Range)**: Q3 - Q1, measures the spread of middle 50%

### What we are going to do:
Calculate quartiles for impressions across dentsu campaigns.

### Why we are doing it:
Quartiles help identify typical campaign reach and detect unusual patterns (outliers).

```python
# METHOD 1: Calculate quartiles using pandas .quantile() function
q1 = df['Impressions'].quantile(0.25)   # 25th percentile
q2 = df['Impressions'].quantile(0.50)   # 50th percentile (median)
q3 = df['Impressions'].quantile(0.75)   # 75th percentile

# Calculate IQR (Interquartile Range)
iqr = q3 - q1

print("QUARTILE ANALYSIS: IMPRESSIONS")
print("=" * 50)
print(f"\nTotal Campaigns: {df['Impressions'].count()}")
print(f"Min Impressions: {df['Impressions'].min():,.0f}")
print(f"Max Impressions: {df['Impressions'].max():,.0f}")
print(f"\nQ1 (25th percentile): {q1:,.0f}")
print(f"Q2 (50th percentile/Median): {q2:,.0f}")
print(f"Q3 (75th percentile): {q3:,.0f}")
print(f"\nIQR (Q3 - Q1): {iqr:,.0f}")

# Identify potential outliers using IQR method
# Outliers are values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr

impressions_data = df['Impressions'].dropna()
outliers = impressions_data[(impressions_data < lower_fence) | (impressions_data > upper_fence)]

print(f"\n" + "-" * 50)
print("OUTLIER DETECTION (1.5 × IQR Rule):")
print("-" * 50)
print(f"Lower Fence: {lower_fence:,.0f}")
print(f"Upper Fence: {upper_fence:,.0f}")
print(f"Number of Outliers: {len(outliers)}")
```

### Code Explanation:
- `df['column'].quantile(0.25)` - Finds the value at the 25th percentile
- `df['column'].quantile(0.50)` - Finds the 50th percentile (same as median)
- IQR = Q3 - Q1 measures the spread of the middle 50% of data
- Outliers are typically defined as values beyond 1.5 × IQR from Q1 or Q3

### Output Explanation:
- Q1 tells us the impressions below which 25% of campaigns fall
- Q2 (median) is the typical campaign reach
- Q3 represents high-reach campaigns
- Outliers are campaigns with unusually high or low impressions

## 8. Percentiles

### What are Percentiles?
A **percentile** indicates the value below which a given percentage of data falls. For example, the 90th percentile is the value below which 90% of the data exists.

### What we are going to do:
Calculate various percentiles for conversion rates in our dentsu campaigns.

### Why we are doing it:
Percentiles help set benchmarks. For example, we can identify the top 10% performing campaigns.

```python
# First, calculate Conversion Rate for each campaign
# Conversion Rate = (Conversions / Clicks) × 100
df_clean = df.dropna(subset=['Conversions', 'Clicks'])
df_clean = df_clean[df_clean['Clicks'] > 0]
df_clean['Conv_Rate'] = (df_clean['Conversions'] / df_clean['Clicks']) * 100

# Calculate various percentiles using .quantile()
p10 = df_clean['Conv_Rate'].quantile(0.10)   # 10th percentile
p25 = df_clean['Conv_Rate'].quantile(0.25)   # 25th percentile
p50 = df_clean['Conv_Rate'].quantile(0.50)   # 50th percentile (median)
p75 = df_clean['Conv_Rate'].quantile(0.75)   # 75th percentile
p90 = df_clean['Conv_Rate'].quantile(0.90)   # 90th percentile
p95 = df_clean['Conv_Rate'].quantile(0.95)   # 95th percentile

print("PERCENTILE ANALYSIS: CONVERSION RATE")
print("=" * 50)
print(f"\nTotal Campaigns: {len(df_clean)}")
print(f"Min Conversion Rate: {df_clean['Conv_Rate'].min():.2f}%")
print(f"Max Conversion Rate: {df_clean['Conv_Rate'].max():.2f}%")

print(f"\n📊 PERCENTILE VALUES:")
print(f"10th Percentile: {p10:.2f}% (Bottom 10% are below this)")
print(f"25th Percentile: {p25:.2f}% (Bottom 25% are below this)")
print(f"50th Percentile: {p50:.2f}% (Median - half above, half below)")
print(f"75th Percentile: {p75:.2f}% (Top 25% are above this)")
print(f"90th Percentile: {p90:.2f}% (Top 10% are above this)")
print(f"95th Percentile: {p95:.2f}% (Top 5% are above this)")

print(f"\n" + "-" * 50)
print("🎯 PERFORMANCE BENCHMARKS:")
print("-" * 50)
print(f"Below Average (< P25): < {p25:.2f}%")
print(f"Average (P25-P75): {p25:.2f}% - {p75:.2f}%")
print(f"Above Average (> P75): > {p75:.2f}%")
print(f"Top Performers (> P90): > {p90:.2f}%")
```

### Code Explanation:
- `df['column'].quantile(0.10)` - Finds the value at the 10th percentile
- `df['column'].quantile(0.90)` - Finds the value at the 90th percentile
- The percentile value tells us: X% of data falls below this value

### Output Explanation:
Percentiles provide clear performance benchmarks:
- Campaigns below the 25th percentile need improvement
- The 50th percentile represents typical performance
- Top performers above the 90th percentile can serve as models

## 9. Skewness

### What is Skewness?
**Skewness** measures the asymmetry of data distribution:
- **Skewness = 0**: Symmetric distribution (like a bell curve)
- **Skewness > 0**: Right-skewed (tail extends to the right, most values on the left)
- **Skewness < 0**: Left-skewed (tail extends to the left, most values on the right)

### What we are going to do:
Calculate skewness of ad spend distribution in our campaign data.

### Why we are doing it:
Understanding data shape helps choose appropriate statistical methods and identify data patterns.

```python
# METHOD 1: Calculate skewness using pandas .skew() function
skewness_spend = df['Ad_Spend_USD'].skew()

# METHOD 2: Understanding skewness conceptually
# If mean > median, data is typically right-skewed
# If mean < median, data is typically left-skewed
mean_spend = df['Ad_Spend_USD'].mean()
median_spend = df['Ad_Spend_USD'].median()

print("SKEWNESS ANALYSIS: AD SPEND DISTRIBUTION")
print("=" * 50)
print(f"\nTotal Campaigns: {df['Ad_Spend_USD'].count()}")
print(f"Mean Ad Spend: ${mean_spend:,.2f}")
print(f"Median Ad Spend: ${median_spend:,.2f}")
print(f"\nSkewness Value: {skewness_spend:.4f}")

# Interpret the result
print("\n📊 INTERPRETATION:")
if skewness_spend > 0.5:
    print(f"→ Right-skewed (positively skewed)")
    print(f"→ Most campaigns have lower ad spend")
    print(f"→ A few campaigns have very high budgets (pulling mean up)")
elif skewness_spend < -0.5:
    print(f"→ Left-skewed (negatively skewed)")
    print(f"→ Most campaigns have higher ad spend")
    print(f"→ A few campaigns have very low budgets")
else:
    print(f"→ Approximately symmetric distribution")
    print(f"→ Ad spend is fairly evenly distributed around the mean")

```

### Code Explanation:
- `df['column'].skew()` - Calculates the skewness of the data
- Skewness > 0.5 is typically considered right-skewed
- Skewness < -0.5 is typically considered left-skewed
- When mean > median, data tends to be right-skewed

### Output Explanation:
The skewness value reveals the shape of budget allocation. A positive skew suggests most campaigns run with moderate budgets while some receive premium investment.

## 10. Kurtosis

### What is Kurtosis?
**Kurtosis** measures the "tailedness" of a distribution - how often extreme values (outliers) occur:
- **Kurtosis = 0**: Normal distribution (mesokurtic)
- **Kurtosis > 0**: Heavy tails, more outliers than normal (leptokurtic)
- **Kurtosis < 0**: Light tails, fewer outliers than normal (platykurtic)

### What we are going to do:
Calculate kurtosis for impressions data.

### Why we are doing it:
High kurtosis warns us about potential extreme values that could skew campaign analysis.

```python
# Calculate kurtosis using pandas .kurtosis() function
# Note: pandas returns "excess kurtosis" (0 for normal distribution)
kurtosis_impressions = df['Impressions'].kurtosis()

print("KURTOSIS ANALYSIS: IMPRESSIONS")
print("=" * 50)
print(f"\nTotal Campaigns: {df['Impressions'].count()}")
print(f"Mean Impressions: {df['Impressions'].mean():,.0f}")
print(f"Std Dev: {df['Impressions'].std():,.0f}")
print(f"\nKurtosis (Excess): {kurtosis_impressions:.4f}")

# Interpret the result
print("\n📊 INTERPRETATION:")
if kurtosis_impressions > 1:
    print(f"→ Leptokurtic (heavy tails)")
    print(f"→ Data has MORE extreme values (outliers) than a normal distribution")
    print(f"→ Some campaigns have exceptionally high or low impressions")
elif kurtosis_impressions < -1:
    print(f"→ Platykurtic (light tails)")
    print(f"→ Data has FEWER extreme values than a normal distribution")
    print(f"→ Impressions are more uniformly distributed")
else:
    print(f"→ Approximately normal (mesokurtic)")
    print(f"→ The distribution of impressions follows a typical pattern")

```

### Code Explanation:
- `df['column'].kurtosis()` - Calculates the excess kurtosis (0 for normal distribution)
- Kurtosis > 1 typically indicates heavy tails (more outliers)
- Kurtosis < -1 typically indicates light tails (fewer outliers)

### Output Explanation:
The kurtosis value indicates whether there are unusual campaigns with exceptionally high or low impressions that might require separate analysis.

## 11. Correlation

### What is Correlation?
**Correlation** measures the strength and direction of the relationship between two variables:
- **Correlation = +1**: Perfect positive relationship (as X increases, Y increases)
- **Correlation = -1**: Perfect negative relationship (as X increases, Y decreases)
- **Correlation = 0**: No linear relationship

Common interpretations:
- |r| > 0.7: Strong correlation
- 0.3 < |r| < 0.7: Moderate correlation
- |r| < 0.3: Weak correlation

### What we are going to do:
Calculate the correlation between ad spend and conversions in dentsu campaigns.

### Why we are doing it:
Understanding if higher ad spend leads to more conversions helps optimize budget allocation.

```python
# METHOD 1: Calculate correlation using pandas .corr() function
correlation = df['Ad_Spend_USD'].corr(df['Conversions'])

# METHOD 2: See correlation matrix for multiple columns
correlation_matrix = df[['Ad_Spend_USD', 'Impressions', 'Clicks', 'Conversions']].corr()

print("CORRELATION ANALYSIS: AD SPEND vs CONVERSIONS")
print("=" * 50)
print(f"\nCorrelation Coefficient: {correlation:.4f}")

# Interpret the correlation
print("\n📊 INTERPRETATION:")
if abs(correlation) > 0.7:
    strength = "Strong"
elif abs(correlation) > 0.3:
    strength = "Moderate"
else:
    strength = "Weak"

direction = "positive" if correlation > 0 else "negative"
print(f"→ {strength} {direction} correlation")

if correlation > 0:
    print(f"→ Higher ad spend tends to be associated with MORE conversions")
else:
    print(f"→ Higher ad spend tends to be associated with FEWER conversions")

# Show full correlation matrix
print("\n" + "-" * 50)
print("FULL CORRELATION MATRIX:")
print("-" * 50)
print(correlation_matrix.round(4))

```

### Code Explanation:
- `df['col1'].corr(df['col2'])` - Calculates correlation between two columns
- `df[['col1', 'col2', 'col3']].corr()` - Creates a correlation matrix for multiple columns
- Correlation ranges from -1 to +1

### Output Explanation:
The correlation coefficient reveals whether spending more leads to more conversions. A positive correlation confirms that budget increases are associated with conversion gains.

## 12. Covariance

### What is Covariance?
**Covariance** measures how two variables change together:
- **Positive covariance**: Variables tend to move in the same direction
- **Negative covariance**: Variables tend to move in opposite directions
- **Zero covariance**: Variables are independent

Unlike correlation, covariance is not standardized, so its value depends on the scale of the data.

### What we are going to do:
Calculate covariance between clicks and conversions.

### Why we are doing it:
Understanding the relationship between clicks and conversions helps dentsu evaluate traffic quality.

```python
# METHOD 1: Calculate covariance using pandas .cov() function
covariance = df['Clicks'].cov(df['Conversions'])

# Also get correlation for comparison (easier to interpret)
correlation = df['Clicks'].corr(df['Conversions'])

print("COVARIANCE ANALYSIS: CLICKS vs CONVERSIONS")
print("=" * 50)
print(f"\nTotal Campaigns: {df['Clicks'].count()}")
print(f"Mean Clicks: {df['Clicks'].mean():,.0f}")
print(f"Mean Conversions: {df['Conversions'].mean():.0f}")
print(f"\nCovariance: {covariance:,.2f}")
print(f"Correlation: {correlation:.4f}")

# Interpret
print("\n📊 INTERPRETATION:")
if covariance > 0:
    print(f"→ Positive covariance")
    print(f"→ As clicks INCREASE, conversions tend to INCREASE")
    print(f"→ More traffic generally leads to more conversions")
elif covariance < 0:
    print(f"→ Negative covariance")
    print(f"→ As clicks increase, conversions tend to decrease")
else:
    print(f"→ Near-zero covariance")
    print(f"→ Clicks and conversions appear to be independent")

# Show covariance matrix
print("\n" + "-" * 50)
print("COVARIANCE MATRIX:")
print("-" * 50)
cov_matrix = df[['Ad_Spend_USD', 'Impressions', 'Clicks', 'Conversions']].cov()
print(cov_matrix.round(2))
```

### Code Explanation:
- `df['col1'].cov(df['col2'])` - Calculates covariance between two columns
- `df[['col1', 'col2']].cov()` - Creates a covariance matrix
- Covariance values can be very large because they're in the original units
- Correlation is easier to interpret because it's always between -1 and +1

### Output Explanation:
Positive covariance between clicks and conversions confirms that more traffic generally leads to more conversions, validating the effectiveness of driving traffic to campaigns.

# Part 2: Inferential Statistics

**Inferential Statistics** allows us to make predictions or inferences about a population based on sample data. This is essential for decision-making when we can't measure everyone.

## 1. Population vs Sample

### What is Population?
A **population** is the entire group we want to study. For example, "all dentsu campaigns ever run" or "all potential customers."

### What is a Sample?
A **sample** is a subset of the population that we actually collect data from. It should be representative of the population.

### What we are going to do:
Demonstrate the difference between population and sample statistics using our campaign data.

### Why we are doing it:
In practice, we rarely have access to entire populations, so we use samples to make inferences.

```python
# Our full dataset represents a "population" of all recorded campaigns
population = df['Conversions'].dropna()

# We'll take a random sample of 50 campaigns
np.random.seed(42)  # This ensures we get the same "random" sample each time
sample = population.sample(n=50)

# Calculate statistics for both
pop_mean = population.mean()
sample_mean = sample.mean()

pop_std = population.std(ddof=0)    # Population std (divide by n)
sample_std = sample.std(ddof=1)     # Sample std (divide by n-1)

print("POPULATION vs SAMPLE COMPARISON")
print("=" * 50)
print(f"\n📊 POPULATION (all {len(population)} campaigns):")
print(f"   Mean Conversions: {pop_mean:.2f}")
print(f"   Std Dev: {pop_std:.2f}")

print(f"\n📋 SAMPLE ({len(sample)} randomly selected campaigns):")
print(f"   Mean Conversions: {sample_mean:.2f}")
print(f"   Std Dev: {sample_std:.2f}")

# Calculate sampling error
sampling_error = abs(pop_mean - sample_mean)
percentage_error = (sampling_error / pop_mean) * 100

print(f"\n⚠️ SAMPLING ERROR:")
print(f"   Difference in Means: {sampling_error:.2f}")
print(f"   Percentage Error: {percentage_error:.2f}%")
print(f"\n💡 This difference is NORMAL and expected!")
print(f"   It's called 'sampling error' - samples are never perfect.")
```

### Code Explanation:
- `population.sample(n=50)` - Randomly selects 50 values from the population
- `np.random.seed(42)` - Sets a "seed" so we get the same random sample each time we run
- `ddof=0` - Divide by n (population formula)
- `ddof=1` - Divide by n-1 (sample formula, which is more accurate for samples)

### Output Explanation:
The sample mean is close but not exactly equal to the population mean. This difference is called **sampling error** and is completely normal.

## 2. Sampling Methods

### What are Sampling Methods?
Different ways to select a sample from a population:

1. **Simple Random Sampling**: Every member has an equal chance of being selected
2. **Systematic Sampling**: Select every nth member from a list
3. **Cluster Sampling**: Divide into clusters, randomly select entire clusters

### What we are going to do:
Demonstrate different sampling methods using our dentsu campaign data.

### Why we are doing it:
Choosing the right sampling method ensures your sample represents the population accurately.

```python
np.random.seed(42)

print("SAMPLING METHODS DEMONSTRATION")
print("=" * 60)
print(f"\nPopulation Size: {len(df)} campaigns")

# ────────────────────────────────────────────────────────────
# 1. SIMPLE RANDOM SAMPLING
# ────────────────────────────────────────────────────────────
simple_sample = df.sample(n=20, random_state=42)

print(f"\n1️⃣ SIMPLE RANDOM SAMPLE (20 campaigns):")
print(f"   Method: df.sample(n=20)")
print(f"   Every campaign has equal chance of being selected")
print(f"   Channels in sample: {simple_sample['Channel'].nunique()} unique channels")

# ────────────────────────────────────────────────────────────
# 2. SYSTEMATIC SAMPLING (every 15th campaign)
# ────────────────────────────────────────────────────────────
systematic_sample = df.iloc[::15]  # Select every 15th row

print(f"\n3️⃣ SYSTEMATIC SAMPLE (every 15th campaign):")
print(f"   Method: df.iloc[::15]")
print(f"   Sample size: {len(systematic_sample)} campaigns")
print(f"   Row indices: 0, 15, 30, 45, ...")

# ────────────────────────────────────────────────────────────
# 3. CLUSTER SAMPLING (select entire regions)
# ────────────────────────────────────────────────────────────
all_regions = df['Region'].unique()
selected_regions = np.random.choice(all_regions, size=2, replace=False)
cluster_sample = df[df['Region'].isin(selected_regions)]

print(f"\n4️⃣ CLUSTER SAMPLE (2 randomly selected regions):")
print(f"   Method: Randomly select entire clusters (regions)")
print(f"   Selected Regions: {list(selected_regions)}")
print(f"   Sample size: {len(cluster_sample)} campaigns")
```

### Code Explanation:
- `df.sample(n=20)` - Simple random sampling
- `df.iloc[::15]` - Systematic sampling (every 15th row)
- `df[df['Region'].isin(selected_regions)]` - Cluster sampling

### Output Explanation:
Each method produces a different sample. Stratified sampling ensures all channels are represented, which is important when channels perform differently.

## 3. Hypothesis Testing Basics

### What is Hypothesis Testing?
**Hypothesis testing** is a statistical method to make decisions based on data. We test whether our data provides enough evidence to reject a claim.

### Key Components:
- **Null Hypothesis (H₀)**: The default assumption (no effect, no difference)
- **Alternative Hypothesis (H₁)**: What we want to prove
- **Significance Level (α)**: Usually 0.05 (5% chance of being wrong)

### What we are going to do:
Explain the concept using a dentsu marketing scenario from our data.

```python
# Calculate the overall average conversion rate
df_clean = df.dropna(subset=['Conversions', 'Clicks'])
df_clean = df_clean[df_clean['Clicks'] > 0]
overall_conv_rate = (df_clean['Conversions'] / df_clean['Clicks'] * 100).mean()

print("HYPOTHESIS TESTING FRAMEWORK")
print("=" * 60)
print(f"""
📋 SCENARIO:
Dentsu wants to test if Search campaigns have a higher 
conversion rate than the overall average of {overall_conv_rate:.2f}%.

────────────────────────────────────────────────────────────
STEP 1: Define Hypotheses
────────────────────────────────────────────────────────────
H₀ (Null Hypothesis): 
   Search campaigns have conversion rate ≤ {overall_conv_rate:.2f}%
   (Search is NOT better than average)

H₁ (Alternative Hypothesis): 
   Search campaigns have conversion rate > {overall_conv_rate:.2f}%
   (Search IS better than average)

────────────────────────────────────────────────────────────
STEP 2: Set Significance Level
────────────────────────────────────────────────────────────
α = 0.05 (5%)

This means we accept a 5% chance of wrongly concluding
Search is better when it actually isn't.

────────────────────────────────────────────────────────────
STEP 3: Decision Rule
────────────────────────────────────────────────────────────
If p-value < 0.05 → Reject H₀ → Search IS better
If p-value ≥ 0.05 → Fail to reject H₀ → Not enough evidence
""")
```

### Code Explanation:
- We calculate the overall average conversion rate as our baseline
- H₀ (null hypothesis) = the default assumption we want to test against
- H₁ (alternative hypothesis) = what we hope to prove
- α (alpha) = significance level, usually 0.05

### Output Explanation:
This framework sets up a structured approach to testing claims with data.

## 4. Type I and Type II Errors

### What are Type I and Type II Errors?

- **Type I Error (False Positive)**: Saying there's an effect when there isn't
- **Type II Error (False Negative)**: Missing a real effect

### What we are going to do:
Illustrate these errors with dentsu's decision-making context.

```python
print("TYPE I AND TYPE II ERRORS")
print("=" * 70)
print("""
                          REALITY
                    ┌─────────────────┬─────────────────┐
                    │   H₀ is TRUE    │   H₀ is FALSE   │
                    │  (No Effect)    │ (Effect Exists) │
             ┌──────┼─────────────────┼─────────────────┤
             │Reject│  TYPE I ERROR   │    CORRECT!     │
   YOUR      │  H₀  │  (False Alarm)  │   (True Pos)    │
  DECISION   ├──────┼─────────────────┼─────────────────┤
             │ Keep │    CORRECT!     │  TYPE II ERROR  │
             │  H₀  │   (True Neg)    │ (Missed Effect) │
             └──────┴─────────────────┴─────────────────┘

────────────────────────────────────────────────────────────────────
🏢 DENTSU EXAMPLE:
────────────────────────────────────────────────────────────────────

Testing: "Does Search channel outperform the average?"

TYPE I ERROR (False Positive):
   → We conclude Search IS better (reject H₀)
   → But actually, it's NOT better
   → CONSEQUENCE: Wasted budget shifting to Search

TYPE II ERROR (False Negative):
   → We conclude Search is NOT better (keep H₀)
   → But actually, it IS better
   → CONSEQUENCE: Missed opportunity for growth

────────────────────────────────────────────────────────────────────
💡 KEY INSIGHT:
────────────────────────────────────────────────────────────────────
• Lower α (stricter) → Fewer Type I errors, MORE Type II errors
• Higher α (lenient) → More Type I errors, FEWER Type II errors
• Larger sample size → Reduces BOTH types of errors
""")
```

### Code Explanation:
This cell explains the two types of errors we can make in hypothesis testing.

### Output Explanation:
Both errors have business consequences. Dentsu should choose α based on which error is more costly for their specific decision.

## 5. p-value

### What is p-value?
The **p-value** is the probability of getting results as extreme as what we observed, IF the null hypothesis were true.

### Simple Interpretation:
- **Small p-value (< 0.05)**: Results are unlikely if H₀ is true → Reject H₀
- **Large p-value (≥ 0.05)**: Results are plausible if H₀ is true → Keep H₀

### What we are going to do:
Calculate a p-value manually to understand how it works.

```python
# Calculate conversion rates for Search vs Other channels
df_clean = df.dropna(subset=['Conversions', 'Clicks'])
df_clean = df_clean[df_clean['Clicks'] > 0]
df_clean['Conv_Rate'] = df_clean['Conversions'] / df_clean['Clicks'] * 100

# Separate Search and Other channels
search_rates = df_clean[df_clean['Channel'] == 'Search']['Conv_Rate']
other_rates = df_clean[df_clean['Channel'] != 'Search']['Conv_Rate']

# Calculate statistics
search_mean = search_rates.mean()
other_mean = other_rates.mean()
search_std = search_rates.std()
search_n = len(search_rates)

# Calculate t-statistic manually
# t = (sample_mean - population_mean) / (sample_std / sqrt(n))
standard_error = search_std / np.sqrt(search_n)
t_statistic = (search_mean - other_mean) / standard_error

print("p-VALUE ANALYSIS: SEARCH vs OTHER CHANNELS")
print("=" * 55)
print(f"\n📊 DATA SUMMARY:")
print(f"   Search Campaigns: {search_n}")
print(f"   Search Mean Conv Rate: {search_mean:.2f}%")
print(f"   Other Channels Mean: {other_mean:.2f}%")
print(f"   Difference: {search_mean - other_mean:.2f}%")

print(f"\n📈 TEST STATISTIC:")
print(f"   Standard Error: {standard_error:.4f}")
print(f"   t-statistic: {t_statistic:.4f}")

# Approximate p-value interpretation
# (In practice, you'd use a t-distribution table or scipy)
print(f"\n📋 INTERPRETATION:")
print(f"   t-statistic = {t_statistic:.2f}")
if abs(t_statistic) > 2:
    print(f"   Since |t| > 2, p-value is likely < 0.05")
    print(f"   → Strong evidence that Search IS different from average!")
else:
    print(f"   Since |t| < 2, p-value is likely > 0.05")
    print(f"   → Not enough evidence to say Search is different")

# Note about p-value
print(f"\n💡 WHAT p-VALUE MEANS:")
print(f"   If Search were NOT better (H₀ true),")
print(f"   what's the probability of seeing this big a difference?")
print(f"   Small p-value = unlikely by chance = real difference!")
```

### Code Explanation:
- We calculate the t-statistic manually: `t = (sample_mean - population_mean) / standard_error`
- Standard error = `std / sqrt(n)` - measures how precise our sample mean is
- Rule of thumb: |t| > 2 usually means p-value < 0.05

### Output Explanation:
The p-value tells us whether the difference we observed is likely due to chance or a real effect.

## 6. Confidence Interval

### What is a Confidence Interval?
A **confidence interval** provides a range of values within which we expect the true population value to fall.

### Simple Explanation:
A 95% confidence interval means: "We're 95% confident the true value is somewhere in this range."

**Formula:** CI = Mean ± (Critical Value × Standard Error)

```python
# Calculate 95% confidence interval for average ad spend
ad_spend = df['Ad_Spend_USD'].dropna()

# Calculate components
n = len(ad_spend)
mean = ad_spend.mean()
std = ad_spend.std()
standard_error = std / np.sqrt(n)

# For 95% confidence, critical value ≈ 1.96 (for large samples)
critical_value = 1.96

# Calculate confidence interval
margin_of_error = critical_value * standard_error
ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

print("CONFIDENCE INTERVAL: AVERAGE AD SPEND")
print("=" * 55)
print(f"\n📊 SAMPLE STATISTICS:")
print(f"   Sample Size (n): {n}")
print(f"   Sample Mean: ${mean:,.2f}")
print(f"   Sample Std Dev: ${std:,.2f}")
print(f"   Standard Error: ${standard_error:,.2f}")

print(f"\n📈 95% CONFIDENCE INTERVAL:")
print(f"   Critical Value (z): {critical_value}")
print(f"   Margin of Error: ${margin_of_error:,.2f}")
print(f"\n   CI = Mean ± Margin of Error")
print(f"   CI = ${mean:,.2f} ± ${margin_of_error:,.2f}")
print(f"\n   ✅ 95% CI: (${ci_lower:,.2f}, ${ci_upper:,.2f})")

print(f"\n💡 INTERPRETATION:")
print(f"   We are 95% confident that the TRUE average ad spend")
print(f"   for all dentsu campaigns is between")
print(f"   ${ci_lower:,.2f} and ${ci_upper:,.2f}")

```

### Code Explanation:
- `standard_error = std / np.sqrt(n)` - Measures precision of our estimate
- `critical_value = 1.96` - For 95% confidence with large samples
- `CI = mean ± (critical_value × standard_error)` - The confidence interval formula

### Output Explanation:
The 95% CI provides a range for expected campaign budgets with a known level of confidence.

## 7. Z-test

### What is a Z-test?
A **Z-test** compares a sample mean to a known population mean when:
1. We know the population standard deviation
2. Sample size is large (n ≥ 30)

**Formula:** Z = (Sample Mean - Population Mean) / (Population Std / √n)

```python
# Z-TEST: Compare Video CTR to industry benchmark

# Calculate CTR for Video campaigns
video_data = df[df['Channel'] == 'Video'].dropna(subset=['Clicks', 'Impressions'])
video_data = video_data[video_data['Impressions'] > 0]
video_ctr = video_data['Clicks'] / video_data['Impressions'] * 100

# Industry benchmark (assumed known values)
industry_mean = 4.0   # Industry average Video CTR
industry_std = 1.5    # Industry standard deviation

# Calculate Z-statistic
sample_mean = video_ctr.mean()
n = len(video_ctr)
z_score = (sample_mean - industry_mean) / (industry_std / np.sqrt(n))

print("Z-TEST: VIDEO CTR vs INDUSTRY BENCHMARK")
print("=" * 55)
print(f"\n📊 DATA:")
print(f"   Video Campaigns: {n}")
print(f"   Dentsu Video CTR: {sample_mean:.2f}%")
print(f"   Industry Mean CTR: {industry_mean}%")
print(f"   Industry Std Dev: {industry_std}%")

print(f"\n📋 HYPOTHESES:")
print(f"   H₀: Dentsu CTR = {industry_mean}% (equals industry average)")
print(f"   H₁: Dentsu CTR ≠ {industry_mean}% (different from industry)")

print(f"\n📈 CALCULATION:")
print(f"   Z = (Sample Mean - Pop Mean) / (Pop Std / √n)")
print(f"   Z = ({sample_mean:.2f} - {industry_mean}) / ({industry_std} / √{n})")
print(f"   Z = {z_score:.4f}")

print(f"\n✅ DECISION (α = 0.05):")
print(f"   Critical values: -1.96 and +1.96")
if abs(z_score) > 1.96:
    print(f"   |Z| = {abs(z_score):.2f} > 1.96 → Reject H₀")
    if sample_mean > industry_mean:
        print(f"\n   🎉 CONCLUSION: Dentsu Video CTR is SIGNIFICANTLY HIGHER")
        print(f"   than the industry average!")
    else:
        print(f"\n   ⚠️ CONCLUSION: Dentsu Video CTR is SIGNIFICANTLY LOWER")
        print(f"   than the industry average.")
else:
    print(f"   |Z| = {abs(z_score):.2f} < 1.96 → Fail to reject H₀")
    print(f"\n   📊 CONCLUSION: No significant difference from industry average.")
```

### Code Explanation:
- `z_score = (sample_mean - industry_mean) / (industry_std / np.sqrt(n))`
- If |Z| > 1.96, we reject H₀ at 95% confidence level
- The Z-score tells us how many standard errors our sample is from the expected value

### Output Explanation:
The Z-test tells us whether dentsu's Video CTR is significantly different from the industry benchmark.

## 8. T-test

### What is a T-test?
A **T-test** is used when:
- Population standard deviation is UNKNOWN
- We use the sample standard deviation instead

### Types of T-tests:
1. **One-Sample T-test**: Compare sample to a known value
2. **Two-Sample T-test**: Compare two different groups

```python
# ONE-SAMPLE T-TEST: Do campaigns meet the 50 conversion target?

conversions = df['Conversions'].dropna()
target = 50  # Target conversions per campaign

# Calculate t-statistic
n = len(conversions)
sample_mean = conversions.mean()
sample_std = conversions.std()
standard_error = sample_std / np.sqrt(n)
t_statistic = (sample_mean - target) / standard_error

print("ONE-SAMPLE T-TEST: MEETING CONVERSION TARGETS")
print("=" * 55)
print(f"\n📊 DATA:")
print(f"   Campaigns: {n}")
print(f"   Sample Mean: {sample_mean:.2f} conversions")
print(f"   Sample Std Dev: {sample_std:.2f}")
print(f"   Target: {target} conversions")

print(f"\n📋 HYPOTHESES:")
print(f"   H₀: Mean conversions = {target} (meeting target)")
print(f"   H₁: Mean conversions ≠ {target} (not meeting target)")

print(f"\n📈 CALCULATION:")
print(f"   t = (Sample Mean - Target) / Standard Error")
print(f"   t = ({sample_mean:.2f} - {target}) / {standard_error:.4f}")
print(f"   t = {t_statistic:.4f}")

# For large samples, use 1.96 as critical value
# For small samples, you'd need a t-table
critical_value = 2.0  # Approximate for large samples

print(f"\n✅ DECISION:")
if abs(t_statistic) > critical_value:
    print(f"   |t| = {abs(t_statistic):.2f} > {critical_value} → Reject H₀")
    if sample_mean > target:
        print(f"\n   🎉 CONCLUSION: Campaigns EXCEED the target!")
    else:
        print(f"\n   ⚠️ CONCLUSION: Campaigns FALL SHORT of target.")
else:
    print(f"   |t| = {abs(t_statistic):.2f} < {critical_value} → Fail to reject H₀")
    print(f"\n   ✅ CONCLUSION: Campaigns are meeting the target.")
```

### Code Explanation:
- **One-sample t-test**: `t = (sample_mean - target) / (sample_std / √n)`


### Output Explanation:
T-tests help us determine if observed differences are statistically significant or just due to random chance.

## 9. Correlation Test

### What is a Correlation Test?
While we calculated correlation earlier, the **correlation test** tells us if the correlation is statistically significant (not due to chance).

### What we are going to do:
Test if the correlation between ad spend and impressions is significant.

```python
# Correlation test: Ad Spend vs Impressions
df_clean = df.dropna(subset=['Ad_Spend_USD', 'Impressions'])
ad_spend = df_clean['Ad_Spend_USD']
impressions = df_clean['Impressions']

# Calculate correlation
r = ad_spend.corr(impressions)
n = len(df_clean)

# Calculate t-statistic for correlation
# t = r × √(n-2) / √(1-r²)
t_statistic = r * np.sqrt(n - 2) / np.sqrt(1 - r**2)

# Calculate R-squared (coefficient of determination)
r_squared = r ** 2

print("CORRELATION TEST: AD SPEND vs IMPRESSIONS")
print("=" * 55)
print(f"\n📊 DATA:")
print(f"   Campaigns: {n}")
print(f"   Correlation (r): {r:.4f}")
print(f"   R-squared (r²): {r_squared:.4f}")

print(f"\n📋 HYPOTHESES:")
print(f"   H₀: ρ = 0 (no correlation exists)")
print(f"   H₁: ρ ≠ 0 (correlation DOES exist)")

print(f"\n📈 TEST:")
print(f"   t-statistic: {t_statistic:.4f}")

# Interpret strength
if abs(r) > 0.7:
    strength = "Strong"
elif abs(r) > 0.3:
    strength = "Moderate"
else:
    strength = "Weak"
direction = "positive" if r > 0 else "negative"

print(f"\n✅ RESULTS:")
if abs(t_statistic) > 2:
    print(f"   |t| = {abs(t_statistic):.2f} > 2 → Correlation IS significant!")
    print(f"\n   📊 INTERPRETATION:")
    print(f"   • {strength} {direction} correlation (r = {r:.2f})")
    print(f"   • R² = {r_squared:.2%} of variation in impressions")
    print(f"     can be explained by ad spend")
    if r > 0:
        print(f"   • Higher ad spend → More impressions ✓")
else:
    print(f"   |t| = {abs(t_statistic):.2f} < 2 → Correlation NOT significant")
    print(f"\n   The relationship might be due to chance.")
```

### Code Explanation:
- `r = ad_spend.corr(impressions)` - Calculate correlation coefficient
- `t = r × √(n-2) / √(1-r²)` - Test if correlation is significant
- `R² = r²` - Proportion of variance explained (coefficient of determination)

### Output Explanation:
A significant correlation confirms that the relationship between ad spend and impressions is real, not just random chance.
