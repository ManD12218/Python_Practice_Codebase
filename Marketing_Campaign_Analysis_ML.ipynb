## 7. Feature Engineering

In this section, we will create new calculated metrics: CTR, Conversion Rate, and ROI.

### 7.1 Creating Click-Through Rate (CTR)

CTR = (Clicks / Impressions) * 100

```python
# Calculating CTR
df['CTR'] = (df['Clicks'] / df['Impressions']) * 100
df['CTR'] = df['CTR'].round(2)

# Displaying result
df[['Campaign_ID', 'Impressions', 'Clicks', 'CTR']].head()
```

**Inference:**

CTR has been calculated successfully. Higher CTR indicates more engaging ad content.

### 7.2 Creating Conversion Rate

Conversion Rate = (Conversions / Clicks) * 100

```python
# Calculating Conversion Rate
df['Conversion_Rate'] = (df['Conversions'] / df['Clicks']) * 100
df['Conversion_Rate'] = df['Conversion_Rate'].round(2)

# Displaying result
df[['Campaign_ID', 'Clicks', 'Conversions', 'Conversion_Rate']].head()
```

**Inference:**

Conversion Rate has been calculated. Higher rates indicate better landing page effectiveness.

### 7.3 Creating ROI

ROI = ((Revenue - Budget) / Budget) * 100

```python
# Calculating ROI
df['ROI'] = ((df['Revenue_Generated'] - df['Budget_Spent']) / df['Budget_Spent']) * 100
df['ROI'] = df['ROI'].round(2)

# Displaying result
df[['Campaign_ID', 'Budget_Spent', 'Revenue_Generated', 'ROI']].head()
```

**Inference:**

ROI has been calculated. Positive values indicate profitable campaigns.

### 7.4 Updated Dataset

Viewing the dataset with new features.

```python
# Displaying updated dataset
df.head()
```

**Inference:**

The dataset now has 11 columns with 3 new features: CTR, Conversion_Rate, and ROI.

### 9.1.1 Weighted Average ROI by Channel

Calculating Weighted Average ROI for each channel.

## 11. Pivot Table Analysis

In this section, we will create pivot tables for multi-dimensional analysis.

### 11.1 Average ROI Pivot

Creating pivot table for average ROI by Channel and Region.

```python
# Pivot table for average ROI
pivot_roi = pd.pivot_table(df, values='ROI', index='Channel', columns='Region', aggfunc='mean')
pivot_roi.round(2)
```

**Inference:**

This matrix shows ROI performance for each channel-region combination.

### 11.1.1 Weighted ROI Pivot

Creating pivot table for Weighted ROI by Channel and Region.

```python
df['ROI_x_Budget_Spent'] = df['ROI'] * df['Budget_Spent']

pivot_roi = (
    pd.pivot_table(df,
                   values=['ROI_x_Budget_Spent','Budget_Spent'],
                   index='Channel',
                   columns='Region',
                   aggfunc='sum')
)
pivot_roi
```

```python
weighted_roi = pivot_roi['ROI_x_Budget_Spent'] / pivot_roi['Budget_Spent']
weighted_roi.round(2)
```

### 11.2 Total Revenue Pivot

Creating pivot table for total revenue by Channel and Region.

```python
# Pivot table for total Revenue
pivot_revenue = pd.pivot_table(df, values='Revenue_Generated', index='Channel', columns='Region', aggfunc='sum')
pivot_revenue.round(2)
```

**Inference:**

This matrix shows revenue contribution for each channel-region combination.

### 11.3 Campaign Count Pivot

Creating pivot table for campaign count by Channel and Region.
