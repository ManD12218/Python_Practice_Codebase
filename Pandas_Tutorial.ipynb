# 1. Importing Libraries

Before using Pandas, we need to import it along with NumPy. The standard convention is to import Pandas as `pd` and NumPy as `np` for convenience.

```python
import numpy as np
import pandas as pd
```

**Output Explanation**: No output is displayed, but both libraries are now ready to use. We can access Pandas functions using `pd.` and NumPy functions using `np.`

# 2. Creating Basic Data Structures

Let's start by creating some basic Python data structures that we'll use to create Pandas objects. We'll create a list of labels, a list of data values, a NumPy array, and a dictionary.

```python
labels = ['a', 'b', 'c']
my_data = [10, 20, 30]
arr = np.array(my_data)
d = {'a': 10, 'b': 20, 'c': 30}

print("Labels:", labels)
print("My data:", my_data)
print("NumPy array:", arr)
print("Dictionary:", d)
```

**Output Explanation**: We've created four different data structures that can all be converted to Pandas Series. This shows the flexibility of Pandas in accepting different input formats.

# PANDAS SERIES

A **Series** is a one-dimensional labeled array. It's like a column in a spreadsheet or a single column of a DataFrame.

---

# 3. Creating a Series from a List

The simplest way to create a Series is from a Python list using `pd.Series()`. Pandas will automatically assign numeric indices (0, 1, 2, ...) if we don't provide custom labels.

```python
pd.Series([10, 20, 30, "@"])
```

**Output Explanation**: The Series shows index values (0, 1, 2, 3) on the left and data values on the right. Notice `dtype: object` because we mixed integers and a string - Pandas converted all to a common type.

---

# 4. Creating a Series with Custom Index

We can provide custom labels (index) for our Series by passing a second argument. This makes data access more meaningful - instead of using numbers, we can use descriptive labels.

```python
pd.Series(my_data, labels)
```

**Output Explanation**: Now the index shows 'a', 'b', 'c' instead of 0, 1, 2. The dtype is `int64` because all values are integers.

---

# 5. Creating a Series from a NumPy Array

We can also create a Series directly from a NumPy array. This is useful when you're working with numerical computations and want to add labels to your data.

```python
pd.Series(arr)
```

**Output Explanation**: The NumPy array is converted to a Series with default numeric indices. The dtype `int32` comes from the NumPy array's data type.

---

# 6. Creating a Series from a Dictionary

When creating a Series from a dictionary, the dictionary keys automatically become the index labels and the values become the data. This is a very common pattern in real-world data processing.

```python
pd.Series(d)
```

**Output Explanation**: Dictionary keys ('a', 'b', 'c') become the index and values (10, 20, 30) become the data. This is very intuitive!

---

# 7. Series with Named Index (US States Example)

Let's create a more practical example - a Series representing some data for US states. We'll use state abbreviations as the index, making the data self-documenting.

```python
ser1 = pd.Series(data=[1, 2, 3, 4], index=['CA', 'OR', 'CO', 'AZ'])
ser1
```

**Output Explanation**: We now have a Series with US state abbreviations as the index. This makes accessing data by state name very intuitive.

---

# 8. Accessing Series Elements by Label

One of the most powerful features of a Series is accessing data using the index label. We can use square bracket notation with the label name, just like accessing dictionary values.

```python
print("Value for California (CA):", ser1['CA'])
print("Value for Arizona (AZ):", ser1['AZ'])
```

**Output Explanation**: We retrieved the values 1 and 4 using their respective state codes. This is much more readable than using numeric indices!

---

# 9. Series Slicing

We can slice a Series just like a list to get a subset of the data. Let's get all elements from index position 1 to the end using Python's slicing syntax.

```python
ser1[1:]
```

**Output Explanation**: The slice `[1:]` returns all elements starting from index position 1 (OR) to the end, giving us OR, CO, and AZ with their values.

---

# 10. Series Arithmetic Operations

When we perform arithmetic operations between two Series, Pandas aligns them by their index labels. If a label exists in both Series, the values are combined. If a label exists in only one Series, the result is `NaN` (Not a Number).

```python
ser1 = pd.Series([1, 2, 3, 4], ['CA', 'OR', 'CO', 'AZ'])
ser2 = pd.Series([1, 2, 5, 4], ['CA', 'OR', 'NV', 'AZ'])

print("Series 1:")
print(ser1)
print("\nSeries 2:")
print(ser2)
```

Now let's add these two Series together. Notice that ser1 has 'CO' while ser2 has 'NV' - these don't match!

```python
ser3 = ser1 + ser2
ser3
```

**Output Explanation**: 
- CA: 1+1=2, OR: 2+2=4, AZ: 4+4=8 (matching indices are added)
- CO: NaN (only in ser1), NV: NaN (only in ser2)
- The dtype changed to `float64` because NaN is a float value

# PANDAS DATAFRAME

A **DataFrame** is a 2-dimensional labeled data structure - like a spreadsheet or SQL table. It has rows and columns, where each column can have a different data type.

---

# 11. Creating a DataFrame from NumPy Array

Let's create a DataFrame from random data. First, we'll set a random seed for reproducibility, then create a 5x4 matrix of random numbers, and finally convert it to a DataFrame with custom row and column labels.

```python
from numpy.random import randn

np.random.seed(10)  # For reproducibility
matrix_data = randn(5, 4)

row_labels = ['A', 'B', 'C', 'D', 'E']
column_headings = ['W', 'X', 'Y', 'Z']

df = pd.DataFrame(data=matrix_data, index=row_labels, columns=column_headings)
df
```

**Output Explanation**: We created a DataFrame with 5 rows (A-E) and 4 columns (W-Z) containing random values from a standard normal distribution. The DataFrame displays like a table with row and column headers.

---

# 12. Checking DataFrame Type

Let's verify that our variable `df` is indeed a Pandas DataFrame by checking its type.

```python
type(df)
```

**Output Explanation**: The output `pandas.core.frame.DataFrame` confirms that `df` is a Pandas DataFrame object.

---

# 13. Selecting Columns

We can select one or more columns from a DataFrame. Use single brackets with a column name to get a Series, or double brackets with a list of column names to get a DataFrame.

```python
# Single column - returns a Series
print("Single column (Series):")
print(df['W'])
print("\nType:", type(df['W']))
```

```python
# Multiple columns - returns a DataFrame
print("Multiple columns (DataFrame):")
df[['X', 'Y']]
```

**Output Explanation**: Single bracket selection returns a Series (1D), while double bracket selection returns a DataFrame (2D). This distinction is important for many operations.

---

# 14. Adding New Columns

We can add new columns to a DataFrame by assigning values to a new column name. The new column can be created from existing columns using arithmetic operations.

```python
# Create a new column from existing columns
df['new'] = df['X'] + df['Y']
df
```

**Output Explanation**: A new column 'new' is added to the DataFrame, containing the sum of columns X and Y for each row.

---

# 15. Dropping Columns

We can remove columns using the `drop()` method. The `axis=1` parameter specifies we're dropping a column (not a row). Use `inplace=True` to modify the original DataFrame, or assign the result to save the change.

```python
# Drop the 'new' column (axis=1 means column)
df = df.drop('new', axis=1)
df
```

**Output Explanation**: The 'new' column has been removed. The `axis=1` parameter tells Pandas to look for columns (axis=0 would be rows).

---

# 16. Dropping Rows

Similarly, we can drop rows by using `drop()` with the row label. By default (axis=0), drop removes rows. Let's see how to drop multiple rows at once.

```python
# Drop rows 'D' and 'A' (returns new DataFrame, doesn't modify original)
df.drop(['D', 'A'])
```

**Output Explanation**: Rows 'D' and 'A' are removed in the returned DataFrame. Note: the original `df` is unchanged unless we use `inplace=True` or reassign.

---

# 17. Selecting Rows with loc and iloc

Pandas provides two main ways to select rows:
- **loc**: Select by label (row name)
- **iloc**: Select by integer position (0, 1, 2, ...)

Let's see both in action.

```python
# Select row 'A' using loc (by label)
print("Using loc (by label):")
print(df.loc['A'])
```

```python
# Select first row using iloc (by position)
print("Using iloc (by position):")
print(df.iloc[0])
```

**Output Explanation**: Both return the same row (row A), but `loc` uses the label name while `iloc` uses the position number. Use `loc` when you know the name, `iloc` when you know the position.

---

# 18. Selecting Specific Cells

We can select specific cells by combining row and column selection using `loc`. The syntax is `df.loc[row, column]` or `df.loc[rows, columns]` for multiple selections.

```python
# Select specific cell: row 'B', column 'Y'
print("Cell at row B, column Y:", df.loc['B', 'Y'])
```

```python
# Select subset: rows A and B, columns W and X
print("Subset of rows and columns:")
df.loc[['A', 'B'], ['W', 'X']]
```

**Output Explanation**: `loc` allows precise selection of specific cells or rectangular subsets of the DataFrame.

---

# 19. Creating DataFrame from Dictionary

A very common way to create a DataFrame is from a dictionary where keys become column names and values (lists) become column data. Let's create a DataFrame with age, weight, and height data.

```python
data = pd.DataFrame({
    'age': [34, 67, 23, 65, 51],
    'weight': [65, 98, 56, 43, 91],
    'height': [160, 180, 172, 166, 180]
}, index=['a', 'b', 'c', 'd', 'e'])

data
```

**Output Explanation**: We created a DataFrame with three columns and five rows (a-e). This dictionary-to-DataFrame conversion is one of the most common patterns in data science.

---

# 20. Conditional Selection (Boolean Filtering)

One of the most powerful features of Pandas is filtering data based on conditions. We can use comparison operators to create boolean masks and filter rows that meet certain criteria.

```python
# Find rows where height > 166
print("People taller than 166cm:")
data[data['height'] > 166]
```

**Output Explanation**: Only rows where height is greater than 166 are returned (b, c, and e).

---

# 21. Multiple Conditions

We can combine multiple conditions using `&` (AND) and `|` (OR) operators. Each condition must be wrapped in parentheses.

```python
# Find rows where height > 166 AND weight > 80
print("People taller than 166cm AND heavier than 80kg:")
data[(data['height'] > 166) & (data['weight'] > 80)]
```

```python
# Select specific columns from filtered data
print("Just the age of people matching criteria:")
data[(data['height'] > 166) & (data['weight'] > 80)]['age']
```

**Output Explanation**: Only rows b and e satisfy both conditions. We can chain column selection to get specific columns from filtered results.

---

# 22. reset_index()

The `reset_index()` method converts the index into a regular column and creates a new numeric index. This is useful when the current index contains important data you want to preserve as a column.

```python
print("Original DataFrame:")
print(data)
print("\nAfter reset_index():")
data.reset_index()
```

**Output Explanation**: The original index (a, b, c, d, e) becomes a new column called 'index', and a new numeric index (0, 1, 2, 3, 4) is created. Use `drop=True` to discard the old index instead of keeping it.

---

# 23. set_index()

The `set_index()` method does the opposite - it takes an existing column and makes it the index. This is useful when a column contains unique identifiers.

```python
# First add a column to use as index
data['name'] = ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve']
print("With name column:")
print(data)
print("\nAfter set_index('name'):")
data.set_index('name')
```

**Output Explanation**: The 'name' column becomes the new index, making it easy to look up data by name.

# HANDLING MISSING DATA

Real-world data often has missing values. Pandas represents missing data as `NaN` (Not a Number). Let's learn how to detect and handle these values.

---

# 24. Creating Data with Missing Values

Let's create a DataFrame that has some missing values (NaN). We'll use `np.nan` to represent missing data.

```python
data_missing = pd.DataFrame({
    'state': ['delhi', 'noida', 'Bengaluru'],
    'cases': [1200, 750, 890],
    'recover': [800, np.nan, 700],
    'death': [80, np.nan, np.nan]
})
data_missing = data_missing.set_index('state')
data_missing
```

**Output Explanation**: We created a DataFrame with COVID-like data. Notice `NaN` values in the 'recover' and 'death' columns - these represent missing data.

---

# 25. Detecting Missing Values with isnull()

The `isnull()` method returns a DataFrame of boolean values showing True where values are missing and False otherwise.

```python
data_missing.isnull()
```

**Output Explanation**: True indicates a missing value. We can see 'noida' has missing values in 'recover' and 'death', and 'Bengaluru' has missing 'death'.

---

# 26. Dropping Rows with Missing Values

The `dropna()` method removes rows that contain any missing values. This is useful when you need complete data for analysis.

```python
data_missing.dropna()
```

**Output Explanation**: Only 'delhi' remains because it's the only row without any missing values. Use `how='all'` to drop rows only if ALL values are missing.

---

# 27. Filling Missing Values with fillna()

Instead of dropping data, we can fill missing values with a specific value, the mean, median, or other statistics. The `fillna()` method is very flexible.

```python
# Fill missing 'death' values with the median of the column
print("Median of death column:", data_missing['death'].median())
print("\nFill with median:")
data_missing['death'].fillna(data_missing['death'].median())
```

```python
# Fill all missing values with 0
print("Fill all NaN with 0:")
data_missing.fillna(0)
```

**Output Explanation**: `fillna()` replaces NaN values. Using median/mean is statistically better than using 0 as it preserves the distribution of data.

# GROUPBY OPERATIONS

The `groupby()` method splits data into groups based on a column, applies a function to each group, and combines the results. It's essential for data aggregation.

---

# 28. Creating Sample Data for GroupBy

Let's create a sales dataset with company names, salesperson names, and sales figures to demonstrate groupby operations.

```python
sales_data = pd.DataFrame({
    'company': ['A', 'B', 'C', 'D', 'B', 'C', 'A', 'A'],
    'person': ['Esh', 'Avi', 'Meg', 'Ram', 'Ash', 'Esh', 'Meg', 'Avi'],
    'sales': [200, 300, 400, 1000, 456, 432, 234, 310]
})
sales_data
```

**Output Explanation**: We have sales data with 8 transactions across 4 companies (A, B, C, D) by various salespeople.

---

# 29. Basic GroupBy - Sum

Let's group the data by company and calculate the total sales for each company.

```python
sales_data.groupby('company')['sales'].sum()
```

**Output Explanation**: Company A has total sales of 744 (200+234+310), B has 756 (300+456), C has 832 (400+432), and D has 1000.

---

# 30. GroupBy with Multiple Aggregations

We can apply multiple aggregation functions at once using `agg()`. This gives us a comprehensive view of each group.

```python
sales_data.groupby('company')['sales'].agg(['sum', 'mean', 'max', 'min', 'count'])
```

**Output Explanation**: For each company, we see the sum, average, maximum, minimum, and count of sales. Company A has 3 transactions (count), D has only 1.

---

# 31. GroupBy with describe()

The `describe()` method provides comprehensive statistics for each group automatically.

```python
sales_data.groupby('company')['sales'].describe()
```

**Output Explanation**: This shows count, mean, std (standard deviation), min, 25% quartile, 50% (median), 75% quartile, and max for each company's sales.

---

# 32. GroupBy with Multiple Columns

We can group by multiple columns to get more detailed breakdowns.

```python
sales_data.groupby(['company', 'person'])['sales'].sum()
```

**Output Explanation**: This shows sales for each person within each company. The result has a multi-level index (company, person).

# 36. Merging DataFrames (SQL-like Join)

The `merge()` function combines DataFrames based on common columns, similar to SQL JOIN. Let's create two DataFrames with a common 'key' column.

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})

print("Left DataFrame:")
print(left)
print("\nRight DataFrame:")
print(right)
```

```python
# Merge on 'key' column
pd.merge(left, right, on='key')
```

**Output Explanation**: The DataFrames are merged on the 'key' column. All columns from both DataFrames are included, with rows matched by their key values.

---

# 37. Different Types of Merges

Like SQL joins, Pandas supports different types of merges:
- **inner**: Only matching keys (default)
- **outer**: All keys from both DataFrames
- **left**: All keys from left DataFrame
- **right**: All keys from right DataFrame

```python
# Create DataFrames with some non-matching keys
left = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                     'A': ['A0', 'A1', 'A2']})

right = pd.DataFrame({'key': ['K1', 'K2', 'K3'],
                      'B': ['B1', 'B2', 'B3']})

print("Left:", list(left['key']))
print("Right:", list(right['key']))

print("\n--- Inner Merge (only matching keys) ---")
print(pd.merge(left, right, on='key', how='inner'))

print("\n--- Outer Merge (all keys) ---")
print(pd.merge(left, right, on='key', how='outer'))

print("\n--- Left Merge (all keys from left) ---")
print(pd.merge(left, right, on='key', how='left'))
```

**Output Explanation**: 
- Inner: Only K1 and K2 (in both)
- Outer: K0, K1, K2, K3 (all keys, NaN for missing values)
- Left: K0, K1, K2 (all from left, NaN for K0's B value)

---

# 38. Joining on Index

We can also join DataFrames based on their indices using the `join()` method or `merge()` with `left_index=True` and `right_index=True`.

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                    index=['K0', 'K1', 'K2'])

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                      'D': ['D0', 'D2', 'D3']},
                    index=['K0', 'K2', 'K3'])

print("Join on index (inner):")
left.join(right, how='inner')
```

**Output Explanation**: Only K0 and K2 are common to both indices, so only those rows appear in the inner join.

# 39. Apply Function

The `apply()` method lets us apply a custom function to each element, row, or column of a DataFrame. This is extremely powerful for data transformation.

```python
# Create a numeric DataFrame
numbers = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [10, 20, 30, 40]
})

print("Original:")
print(numbers)

# Apply a function to each element
print("\nSquare each element:")
print(numbers.apply(lambda x: x**2))
# NOTE : print(numbers.apply(lambda x: x['A']*x['B'])) - throws error because in that case x would be column A in 
# first iteration of lambda function since x will be column A since axis =0 since not mentioned explicitly here and the next step would be
# to find A['A'] which doesnot exists. Instead we need to mention explicitly that axis = 1 whcih would then make it row01['A']*rw01['B'] whcih makes sense.
```

```python
# Apply function to each column
print("Sum of each column:")
numbers.apply(sum)
```

```python
# Apply function to each row
print("Sum of each row:")
numbers.apply(sum, axis=1)
```

**Output Explanation**: `apply()` is versatile - use `axis=0` (default) to apply to columns, `axis=1` for rows. Lambda functions make it easy to write quick transformations.

# 40. Common DataFrame Methods

Let's look at some commonly used DataFrame methods for quick data exploration.

```python
# Create sample data
sample = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'age': [25, 30, 35, 28, 32],
    'salary': [50000, 60000, 75000, 55000, 65000],
    'department': ['HR', 'IT', 'IT', 'HR', 'Finance']
})

print("DataFrame:")
print(sample)
```

```python
# head() - View first n rows
print("First 2 rows (head):")
print(sample.head(2))
```

```python
# tail() - View last n rows
print("Last 2 rows (tail):")
print(sample.tail(2))
```

```python
# info() - DataFrame information
print("DataFrame info:")
sample.info()
```

```python
# describe() - Statistical summary
print("Statistical summary:")
sample.describe()
```

```python
# value_counts() - Count unique values
print("Department counts:")
print(sample['department'].value_counts())
```

```python
# unique() and nunique() - Get unique values
print("Unique departments:", sample['department'].unique())
print("Number of unique departments:", sample['department'].nunique())
```

**Output Explanation**: These methods are essential for initial data exploration. `head()`/`tail()` for quick looks, `info()` for structure, `describe()` for statistics, `value_counts()` for distributions.

# 41. Sorting Data

We can sort DataFrames by values in one or more columns using `sort_values()`.

```python
# Sort by salary (ascending)
print("Sorted by salary (ascending):")
print(sample.sort_values('salary'))
```

```python
# Sort by salary (descending)
print("Sorted by salary (descending):")
print(sample.sort_values('salary', ascending=False))
```

```python
# Sort by multiple columns
print("Sorted by department, then by age:")
print(sample.sort_values(['department', 'age']))
```

**Output Explanation**: `sort_values()` reorders rows. Use `ascending=False` for descending order. Pass a list of columns for multi-level sorting.
